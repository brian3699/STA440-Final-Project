### Finding the calorie consumption 3 hour prior to sleep       

```python
import pandas as pd
from datetime import datetime, timedelta

file_path = 'minuteSleep_merged.csv'
sleep_data = pd.read_csv(file_path)
file_path_minute_calories = 'minuteCaloriesNarrow_merged.csv'
minute_calories_data = pd.read_csv(file_path_minute_calories)


# Function to convert time in sleep_summary to datetime object
def convert_to_datetime(date_str, time_str):
    datetime_str = date_str + ' ' + time_str
    return datetime.strptime(datetime_str, '%m/%d/%y %m/%d/%y %H:%M')

# Function to calculate total calories burned 3 hours before sleep
def calculate_calories_3hr_before_sleep(row, calorie_data):
    start_time = convert_to_datetime(row['date'], row['start_time'])
    start_time_3hr_before = start_time - timedelta(hours=3)
    filtered_data = calorie_data[(calorie_data['Id'] == row['id']) & 
                                 (calorie_data['ActivityMinute'] >= start_time_3hr_before.strftime('%m/%d/%Y %H:%M:%S %p')) & 
                                 (calorie_data['ActivityMinute'] < start_time.strftime('%m/%d/%Y %H:%M:%S %p'))]
    return filtered_data['Calories'].sum()

minute_calories_data['ActivityMinute'] = pd.to_datetime(minute_calories_data['ActivityMinute'])

sleep_data['calories_3hour_before_sleep'] = sleep_data.apply(lambda row: calculate_calories_3hr_before_sleep(row, minute_calories_data), axis=1)

output_file_path = 'sleep.csv'
sleep_data.to_csv(output_file_path, index=False)

```

### Calculating the calories_daytime

```python
mport pandas as pd
from datetime import datetime, timedelta

file_path = 'sleep.csv'
sleep_data = pd.read_csv(file_path)
sleep_data = sleep_data.dropna()
sleep_data['wake_time'] = pd.to_datetime(sleep_data['wake_time'])
sleep_data['start_time'] = pd.to_datetime(sleep_data['start_time'])

#file_path_minute_calories = '/Users/younghoojun/Desktop/[440] Final Ind Case Study/minuteCaloriesNarrow_merged.csv'
#minute_calories_data = pd.read_csv(file_path_minute_calories)

def parse_date_time(date_str, time_str):
    return datetime.strptime(time_str, '%m/%d/%y %H:%M')



def calculate_calories_daytime(row, minute_calories_data):
    wake_time =  row['wake_time']
    start_time = row['start_time']

    
    start_range = wake_time + timedelta(hours=3)
    end_range = start_time - timedelta(hours=3)

    duration = int((end_range - start_range).total_seconds() / 60)

    filtered_data = minute_calories_data[
        (minute_calories_data['Id'] == row['id']) &
        (minute_calories_data['ActivityMinute'] >= start_range) &
        (minute_calories_data['ActivityMinute'] < end_range)
    ]

    return filtered_data['Calories'].mean()*duration

minute_calories_data['ActivityMinute'] = pd.to_datetime(minute_calories_data['ActivityMinute'])

sleep_data['calories_daytime'] = sleep_data.apply(
    lambda row: calculate_calories_daytime(row, minute_calories_data), axis=1)

output_file_path = 'sleep.csv'
sleep_data.to_csv(output_file_path, index=False)

output_file_path
```
